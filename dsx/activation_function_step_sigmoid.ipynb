```python
import numpy as np
import matplotlib.pyplot as plt

# ステップ関数
def step_function(x):
    return np.array(x > 0, dtype=np.int)

# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 関数計算
x = np.arange(-10.0, 10.0, 0.1)
y1 = step_function(x)
y2 = sigmoid(x)

# グラフ描画
plt.plot(x,y1, linestyle="--")
plt.plot(x,y2)
plt.ylim(-0.1, 1.1)
plt.show()
```
